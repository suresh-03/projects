```python
class user:
    def __init__(self, name):
        self.name = name
        self.login_attempts = 0

    def increment_login_attempts(self):
        self.login_attempts += 1
        print("login attempts updated.")

    def details(self):
        print("username:", self.name)
        print("login attempts:", self.login_attempts)

    def reset(self):
        self.login_attempts = 0
        print("login attempts reseted to default.")


user_name = input("enter your name:")
user1 = user(user_name)
while True:
    login = input("want to login type 'yes' or \n type 'no':")
    login = login.upper()
    if login == "YES":
        user1.increment_login_attempts()
    elif login == "NO":
        print("login cancelled.")
        break
    else:
        print("give valid response!")
        break
print()
user1.details()
print()
reset_login = input("want to reset login attempts type 'yes' or 'no':")
reset_login = reset_login.upper()
if reset_login == "YES":
    user1.reset()
elif reset_login == "NO":
    print("reset cancelled.")
else:
    print("give valid response!")
print()
user1.details()
```

```python
class restaurent:
    def __init__(self, name, Type):
        self.name = name
        self.Type = Type
        self.open_time = "9.00 am"
        self.close_time = "10.00 pm"

    def describe(self):
        print("restaurent name:", self.name)
        print("restaurent Type:", self.Type)
        print("opening Time:", self.open_time)
        print("closing Time:", self.close_time)


class iceCreamStand(restaurent):
    def __init__(self, name, Type):
        super().__init__(name, Type)
        self.flavors = ["chocolate", "strawberry", "butterscotch", "blackcurrent"]

    def display_flavors(self):
        print("ice cream flavors available:\n")
        i = 1
        for flavor in self.flavors:
            print("\t\t\t", str(i) + "." + flavor)
            i = i + 1


ice = iceCreamStand("SSS restaurent", "veg and non-veg")
ice.describe()
print()
ice.display_flavors()
```

```python
class car:
    def __init__(self, name, model):
        self.name = name
        self.model = model

    def describe(self):
        print("carname:", self.name)
        print("carmodel:", self.model, "\n")


class ElectricCar(car):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.car_upgrade = Battery()


class Battery:
    def __init__(self):
        self.battery_size = 200
        self.up_battery = 100

    def describe_battery(self):
        print("the battery capacity is ", str(self.battery_size) + "-KWH")

    def upgdrade_battery(self):
        self.battery_size += self.up_battery
        print("the battery in the car is upgraded!\n")

    def range_car(self):
        if self.battery_size == 200:
            self.range = 300
        elif self.battery_size == 300:
            self.range = 410

        print("In full charge the car can go up to", str(self.range) + "KM\n")


user = ElectricCar("tesla", "model S")
user.describe()
user.car_upgrade.describe_battery()
user.car_upgrade.range_car()
user.car_upgrade.upgdrade_battery()
user.car_upgrade.describe_battery()
user.car_upgrade.range_car()
```
